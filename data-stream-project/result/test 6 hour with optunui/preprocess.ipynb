```python
import pandas as pd

# Replace 'example.csv' with the path to your CSV file
csv_file_path = '../../river_radar/merged_data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
print(df.head(10))
```

```python
# prompt: in the first column is a datetime rename into hour, and only pick the hour from the datetime, shift the column hour to first column and delete datetimeUTC

df['hour'] = pd.to_datetime(df['datetimeUTC']).dt.hour
df = df[['hour'] + list(df.columns[:-1])]
df = df.drop('datetimeUTC', axis=1)
print(df.head(10))
```

```python
# Shift column Tairua to end
cols = list(df.columns)
cols.remove('Tairua')
cols.append('Tairua')
df = df[cols]

print(df.head())
```

```python
# prompt: if exists column Tairua, change the name into Water Level

if 'Tairua' in df.columns:
  df = df.rename(columns={'Tairua': 'Water Level'})

print(df.head(10))
```

```python
# prompt: create new column called Water Movement where Water Movement row n = Water Level row n+24 - Water Level row n

df['Water Movement in 6 Hour'] = df['Water Level'].shift(-24) - df['Water Level']
df.head(100)
```

```python
# prompt: delete the last 25 row

df = df[:-25]
```

```python
# prompt: save to csv with name preprocess

df.to_csv('preprocess.csv', index=False, header=False)
```

```python
columns_to_process = df.columns[1:-5]  # Exclude the first and last columns
df[columns_to_process] = (df[columns_to_process]+20) / 20
df['hour'] = df['hour']/24
df.head(5)
```

```python
df.to_csv('preprocessNormalized.csv', index=False, header=False)
```
