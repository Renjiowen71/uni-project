```python
import pandas as pd

# Replace 'example.csv' with the path to your CSV file
csv_file_path = '../../river_radar/merged_data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

if 'Opitonui' in df.columns:
    df = df.drop('Opitonui', axis=1)

df['hour'] = pd.to_datetime(df['datetimeUTC']).dt.hour
df = df[['hour'] + list(df.columns[:-1])]
df = df.drop('datetimeUTC', axis=1)

df['Water Movement in 6 Hour'] = df['Tairua'].shift(-24) - df['Tairua']
df.head(100)

```

```python
df_80_percent = df.iloc[:60000]
df_20_percent = df.iloc[60000:86000]

df_duplicated_80_percent = pd.concat([df_80_percent] * 3, ignore_index=True)
df_duplicated_80_percent_shuffled = df_duplicated_80_percent.sample(frac=1).reset_index(drop=True)

df = pd.concat([df_duplicated_80_percent_shuffled, df_20_percent], ignore_index=True)

# Print the shape of the new DataFrame
print(f"Shape of the duplicated DataFrame: {df.shape}")

# Optionally, you can save the duplicated DataFrame to a CSV file
df.to_csv("duplicatedMergedData.csv", index=False)
```

```python
# prompt: if exists column Tairua, change the name into Water Level

if 'Tairua' in df.columns:
  df = df.rename(columns={'Tairua': 'Water Level'})

print(df.head(10))
```

```python
columns_to_process = df.columns[1:-4]  # Exclude the first and last columns
df[columns_to_process] = (df[columns_to_process]+20) / 30
df['hour'] = df['hour']/24
df.head(5)
```

```python
df.to_csv("preprocessNormalizedDuplicated.csv", index=False, header=False)
```
