```python
import pandas as pd

# Replace 'example.csv' with the path to your CSV file
csv_file_path = 'river_radar/merged_data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

print(df.iloc[:10])
```

```python
# Delete the first column
df = df.iloc[:, 1:]

# Display the new DataFrame
print(df.iloc[:10])
print(df.columns)
```

```python
# Delete the column with name "Opitonui"
df = df.drop(columns=['Opitonui'])

# Display the new DataFrame without the "Opitonui" column
print(df.iloc[:10])
```

```python
# Extract the target variable 'tairua'
target = df['Tairua']
target = (target * 10).astype(int)
# Remove the 'tairua' column from the DataFrame
features = df.drop(columns=['Tairua'])
print("Target variable (Tairua):")
print(target)
print("\nFeatures (without Tairua column):")
print(features)
```

```python
import matplotlib.pyplot as plt
# Plot histograms for the first 10 features
for column in features.columns[:10]:
    plt.figure()  # Create a new figure for each histogram
    plt.hist(features[column], bins=10, color='skyblue', edgecolor='black')
    plt.xlabel('Normalized Values')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
# Calculate the maximum value for every column
max_values = features.max()

# Display the maximum value for every column
print(max_values)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize every column
normalized_df = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)

# Display the normalized DataFrame
print(normalized_df)
```

```python
# Plot histograms for the first 10 features
for column in normalized_df.columns[:10]:
    plt.figure()  # Create a new figure for each histogram
    plt.hist(normalized_df[column], bins=10, color='skyblue', edgecolor='black')
    plt.xlabel('Normalized Values')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
combined_df = pd.concat([normalized_df, target], axis=1)

# Specify the file path relative to the current directory
output_csv_path = 'river_radar/modified_data2.csv'

# Save the combined DataFrame to a CSV file
combined_df.to_csv(output_csv_path, index=False)
```
