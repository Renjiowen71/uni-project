```python
import pandas as pd

# Replace 'example.csv' with the path to your CSV file
csv_file_path = 'river_radar/merged_data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
df.head(10)
```

```python
# prompt: in the first column is a datetime rename into hour, and only pick the hour from the datetime, shift the column hour to first column and delete datetimeUTC

df['hour'] = pd.to_datetime(df['datetimeUTC']).dt.hour
df = df[['hour'] + list(df.columns[:-1])]
df = df.drop('datetimeUTC', axis=1)
print(df.head(10))
```

```python
# prompt: erase if existcolumn Opitonui and shift column Tairua to end

if 'Opitonui' in df.columns:
    df = df.drop('Opitonui', axis=1)

# Shift column Tairua to end
cols = list(df.columns)
cols.remove('Tairua')
cols.append('Tairua')
df = df[cols]

print(df.head())
```

```python
# prompt: if exists column Tairua, change the name into Water Level

if 'Tairua' in df.columns:
  df = df.rename(columns={'Tairua': 'Water Level'})

print(df.head(10))
```

```python
# prompt: create new column called Water Movement where Water Movement row n = Water Level row n+16 - Water Level row n

df['Water Movement in 4 Hour'] = df['Water Level'].shift(-16) - df['Water Level']
df.head(17)
```

```python
# prompt: delete the last 20 row

df = df[:-20]
```

```python
# prompt: save to csv with name preprocess

df.to_csv('river_radar/preprocess.csv', index=False)
```
